```python
import pandas as pd

# 데이터프레임 생성
data = [
    ["1,000", "1,100", '1,510'],
    ["1,410", "1,420", '1,790'],
    ["850", "900", '1,185'],
]
columns = ["03/02", "03/03", "03/04"]
df = pd.DataFrame(data=data, columns=columns)
df.info()
```

```python
# 쉼표 제거하고 Int 로 변환하는 함수 정의
# 이렇게 작업을 한 후에 함수 만드는 것임 / 바로 함수 만드는게 아니라
x = '1,000'
x.replace(',','')
```

```python
def rm_comma(x):
    return int(x.replace(',',''))
```

```python
df['03/02'] = df['03/02'].apply(rm_comma)
df['03/03'] = df['03/03'].apply(rm_comma)
df.info()
```

## matplotlib

```python
import matplotlib.pyplot as plt

dates = [
    '2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05',
    '2021-01-06', '2021-01-07', '2021-01-08', '2021-01-09', '2021-01-10'
]
min_temperature = [20.7, 17.9, 18.8, 14.6, 15.8, 15.8, 15.8, 17.4, 21.8, 20.0]
max_temperature = [34.7, 28.9, 31.8, 25.6, 28.8, 21.8, 22.8, 28.4, 30.8, 32.0]
```

```python
import matplotlib.pyplot as plt

# 날짜를 x축으로 사용하기 위해 변환
plt.figure(figsize=(10, 5))

# 최소 기온 그래프
plt.plot(dates, min_temperature, label='Min Temperature', color='blue', marker='o')

# 최대 기온 그래프
plt.plot(dates, max_temperature, label='Max Temperature', color='red', marker='o')

# 그래프 제목과 축 라벨 추가
plt.title('Daily Min and Max Temperatures')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')

# x축 라벨 회전 (가독성을 위해)
plt.xticks(rotation=45)

# 범례 추가
plt.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 데이터 준비
dates = [
    '2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05',
    '2021-01-06', '2021-01-07', '2021-01-08', '2021-01-09', '2021-01-10'
]
min_temperature = [20.7, 17.9, 18.8, 14.6, 15.8, 15.8, 15.8, 17.4, 21.8, 20.0]
max_temperature = [34.7, 28.9, 31.8, 25.6, 28.8, 21.8, 22.8, 28.4, 30.8, 32.0]

# Figure와 Axes 생성
fig, ax = plt.subplots(figsize=(10, 5))

# 최소 기온 그래프
ax.plot(dates, min_temperature, label='Min Temperature', color='blue', marker='o')

# 최대 기온 그래프
ax.plot(dates, max_temperature, label='Max Temperature', color='red', marker='o')

# 제목과 라벨 설정
ax.set_title('Daily Min and Max Temperatures')
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (°C)')

# x축 라벨을 회전하여 가독성 향상
ax.set_xticks(range(len(dates)))
ax.set_xticklabels(dates, rotation=45)

# 범례 추가
ax.legend()

# 레이아웃 자동 조정
fig.tight_layout()

# 그래프 표시
plt.show()
```

```python
import matplotlib.pyplot as plt

dates = [
    '2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04', '2021-01-05',
    '2021-01-06', '2021-01-07', '2021-01-08', '2021-01-09', '2021-01-10'
]
min_temperature = [20.7, 17.9, 18.8, 14.6, 15.8, 15.8, 15.8, 17.4, 21.8, 20.0]
max_temperature = [34.7, 28.9, 31.8, 25.6, 28.8, 21.8, 22.8, 28.4, 30.8, 32.0]

# artist layer(ax) , figure layer(fig) 따로 생성
# 하나하나가 다 독립적인 instance 임 
fig, ax = plt.subplots(nrows =1, ncols =1, figsize=(10,6))
ax.plot(dates, min_temperature, label ='Min_Temp')
ax.plot(dates, max_temperature, label ='Max_Temp')
ax.legend()
plt.show()
```

```python
print(fig) # 그래프 자체 크기
print(ax) # 그래프 내부 line을 의미함
```
